```python
import numpy as np
from scipy.stats import binom
from scipy.optimize import brentq

def calltree_numpy(S0, K, T, r, sigma, N, q):
    deltaT = T / float(N)
    u = np.exp(sigma*np.sqrt(deltaT))
    d = 1.0 / u
    p = (np.exp((r-q)*deltaT)-d) / (u-d)
    piu = np.exp(-r*deltaT) * p
    pid = np.exp(-r*deltaT) * (1-p)
    C = np.zeros((N+1, N+1))
    
    # Stock price
    S = S0 * u**np.arange(N+1) * d**(2*np.arange(N+1)[:, np.newaxis])
    S = np.triu(S)  #keep only the upper triangular part

    # Final payoffs
    C[:, N] = np.maximum(0, S[:, N]-K)  #note: np.maximum in place of max
    
    # Work backwards through the tree
    for j in range(N-1, -1, -1):
        C[:j+1, j] = piu * C[:j+1, j+1] + pid * C[1:j+2, j+1]
        
    return  C[0, 0]
S0=12.; K=15.; T=3/12.; r=.02; sigma=.25; q= 0.01; N=500;
print(calltree_numpy(S0, K, T, r, sigma, N, q))
```

```python
from numba import jit
@jit(nopython=True)
def chooser(S0, K, T, r, sigma, N, M):
    # Parameters
    deltaT = T / float(N)
    u = np.exp(sigma*np.sqrt(deltaT))
    d = 1.0 / u
    p = (np.exp((r)*deltaT)-d) / (u-d)
    piu = np.exp(-r*deltaT) * p
    pid = np.exp(-r*deltaT) * (1-p)
    C = np.zeros((N+1, N+1))
    S = np.zeros((N+1, N+1))
    P = np.zeros((N+1, N+1))
    
    # Stock price tree
    for i in range(N+1):
        for j in range(i, N+1):
            S[i, j] = S0 * u**j * d**(2*i)
            
    # Determine final payoffs
    for i in range(N+1):
        C[i, N] = max(0, S[i, N]-K)
        P[i, N]= max(0, K-S[i, N])        
        
    # Compute call price tree
    for j in range(N-1, -1, -1):
        for i in range(j+1):
            C[i, j] = piu * C[i, j+1] + pid * C[i+1, j+1]
            P[i, j] = piu * P[i, j+1] + pid * P[i+1, j+1]
    CH = np.zeros((M+1,M+1))
    for i in range(M+1):
        CH[i, M] = max(C[i, M], P[i, M])
    for j in range(M-1, -1, -1):
        for i in range(j+1):
            CH[i, j] = piu * CH[i, j+1] + pid * CH[i+1, j+1]
    
    return CH[0, 0]
S0=12.; K=15.; T=3/12.; r=.02; sigma=.25; q= 0.01; N=500; M=200
print(chooser(S0, K, T, r, sigma, N, M))
%timeit chooser(S0, K, T, r, sigma, N, M)
```
